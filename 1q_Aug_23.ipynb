```python
#finding the missing values
df.isnull().sum()
```

```python
# Sum of cases

Total_cases = df.Confirmed.sum()
Total_cases
```

```python
Total_deaths = df.Deaths.sum()
Total_deaths
```

```python
Total_Recovered = df.Recovered.sum()
Total_Recovered
```

```python
#calculating death rate
death_rate = Total_deaths/Total_cases * 100
death_rate
```

```python
#calculating recovery rate
recovered_rate = Total_Recovered/Total_cases * 100
recovered_rate
```

```python
#vusialize the nan value in heatmap

sns.heatmap(df.isnull(),yticklabels=False, cbar=False, cmap='viridis')
```

```python
# What is the mode of "Travel_history"
print('mode is:', df['Travel_history'].mode()[0])
```

```python
# visualize & explore

sns.countplot(data=df, x='Country/Region')
plt.xticks(rotation=90)
```

```python
sns.pairplot(data=df, hue='Country/Region')
```

```python
#pairplot
sns.pairplot(data=df, vars=['Country/Region', 'Confirmed', 'Deaths', 'Recovered', 'Active',
       'New cases', 'New deaths', 'New recovered', 'Deaths / 100 Cases', 'Recovered / 100 Cases', 'Deaths / 100 Recovered',
       'Confirmed last week', '1 week change', '1 week % increase',
       'WHO Region'])
```

```python
#swarm plot
sns.swarmplot(data=df, y='Country/Region', x='Confirmed')
```

```python
#barplot

sns.barplot(data=df, y='Confirmed',x='Country/Region', ci=True)
plt.xticks(rotation=90)
```
